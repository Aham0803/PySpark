create a DataFrame in PySpark

```python
employee_data =[
    (100,"gaurav yadav" , "IT" , 10000 , "Delhi"),
    (101,"saurab yadav" , "HR" , 10000 , "chennai"),
    (102,"dev yadav" , "IT" , 10000 , "Delhi"),
    (103,"arpit yadav" , "Finance" , 10000 , "Banglore"),
    (104,"amit yadav" , "IT" , 10000 , "vadodra"),
    (105,"rohit yadav" , "Finance" , 10000 , "Ahmedabad"),
    (106,"khusi yadav" , "IT" , 10000 , "surat"),
    (107,"divya yadav" , "HR" , 10000 , "up")
]

columns = ["emp_id" , "emp_name" , "department","salary" ,"location"]
```

```python
df = spark.createDataFrame(data = employee_data , schema = columns)
print("\n original DataFrame")
df.show()
df.printSchema()
```
