starting the pyspark session

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("demo").getOrCreate()
```

create a dataframe in pyspark

```python
df = spark.createDataFrame(
    [
        ("sue" , 32),
        ("li" , 3),
        ("bob" , 75),
        ("heo" , 13)
    ],
    ["first_name","age"],
)
```
