```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
```

```python
spark = SparkSession.builder.appName("conditionalAgg").getOrCreate()
```

```python
df = spark.read.csv("/content/Automobile_data.csv" , header = True , inferSchema= True)
```

```python
df.show(5)
```

```python
df_conditional = df.groupBy("fuel-type").agg(
    F.sum(F.when(F.col("length")>0 , F.col("length"))).alias("TotalLength"),
    F.count("*").alias("bodies"),
    F.count(F.when(F.col("num-of-doors") =="two" , 1)).alias("two_door"),
    F.count(F.when(F.col("num-of-doors") =="four" , 1)).alias("four_door")
)
```
